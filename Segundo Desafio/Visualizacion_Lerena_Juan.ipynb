## Distribucion de la edad en el total de la muestra y por estado (activo y baja)



```python
fig , ax = plt.subplots(figsize=(10,10),nrows=3)
ax[0].hist(bank["Customer_Age"], bins=47)
ax[1].hist(bank.loc[bank["Attrition_Flag"]=="Existing Customer"]["Customer_Age"], bins=47)
ax[2].hist(bank.loc[bank["Attrition_Flag"]=="Attrited Customer"]["Customer_Age"], bins=21)
ax[0].set_title("Total Muestra")
ax[1].set_title("Clientes Activos")
ax[2].set_title("Clientes Baja")

```

# Cantidad de dependencias

```python
conteo_categ = bank["Dependent_count"].value_counts().sort_index()

fig , ax = plt.subplots()
ax.bar(conteo_categ.index, conteo_categ.values)
```

# Distribucion sobre antiguedad


```python
fig , ax = plt.subplots(figsize=(12,7), nrows=2)
ax[0].hist(bank.loc[bank["Attrition_Flag"]=="Existing Customer"]["Antiquity_months"], bins=15)
ax[0].set_title("Antiguedad (meses) de los activos")
ax[1].hist(bank.loc[bank["Attrition_Flag"]=="Attrited Customer"]["Antiquity_months"], bins=15)
ax[1].set_title("Antiguedad (meses) de los baja")
```
